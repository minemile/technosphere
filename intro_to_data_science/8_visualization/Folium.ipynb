```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import folium
from folium.plugins import MarkerCluster
%matplotlib inline
```

```python
data = pd.read_csv('merchants_train.csv', sep=';')
```

```python
mos_center = [55.750475, 37.619020]
mos_min_lat, mos_max_lat = 55.564141, 55.908930
mos_min_lon, mos_max_lon = 37.344362, 37.879945
```

```python
data_moscow = data[data.latitude.between(mos_min_lat, mos_max_lat) &
                   data.longitude.between(mos_min_lon, mos_max_lon)]
```

```python
data.shape, data_moscow.shape
```

```python
moscow = folium.Map(location=mos_center, zoom_start=11, control_scale=True)
```

```python
moscow
```

```python
random_places = data_moscow[['latitude', 'longitude']].sample(200)
random_places.shape
```

```python
# Не стройте сразу очень много точек, folium не выдержит
marker_cluster = MarkerCluster(locations=random_places[['latitude', 'longitude']].values)
```

```python
marker_cluster.add_to(moscow)
```
