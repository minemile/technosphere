```python
import pandas as pd
import numpy as np
```

```python
salaries_by_region = pd.read_csv('salaries-by-region.csv')
salaries_by_college_type = pd.read_csv('salaries-by-college-type.csv')
for col in salaries_by_region.columns[2:]:
    salaries_by_region[col] = salaries_by_region[col].str.strip('$').str.replace(',', '').astype(float)
    
for col in salaries_by_college_type.columns[2:]:
    salaries_by_college_type[col] = salaries_by_college_type[col].str.strip('$').str.replace(',', '').astype(float)
```

```python
salaries_by_region['Starting Median Salary'] = salaries_by_region['Starting Median Salary'].replace( '[\$,)]','', regex=True).astype('float32')
```

```python
salaries_by_college_type['Starting Median Salary'] = salaries_by_college_type['Starting Median Salary'].replace( '[\$,)]','', regex=True).astype('float32')
```

```python
salaries_by_college_type[salaries_by_college_type['School Type'] == 'Engineering'].sort_values(by='Starting Median Salary')
```

```python
salaries_by_college_type.groupby(['School Type'])['Starting Median Salary'].mean().round()
```

```python
salaries_by_region['Mid-Career 90th Percentile Salary'] = salaries_by_region['Mid-Career 90th Percentile Salary'].str.strip('$').str.replace(',', '').astype(float) 
```

```python
aggs = {"Mid-Career 90th Percentile Salary": ["mean", 'max', 'min', 'std']}
salaries_by_region.groupby('Region').agg(aggs).round()
```

```python
salaries_by_college_type['Mid-Career 25th Percentile Salary'] = salaries_by_college_type['Mid-Career 25th Percentile Salary'].replace( '[\$,)]','', regex=True).astype('float32')
```

```python
salaries_by_region['Mid-Career 25th Percentile Salary'] = salaries_by_region['Mid-Career 25th Percentile Salary'].replace( '[\$,)]','', regex=True).astype('float32')
```

```python
joined_salaries = salaries_by_college_type.merge(salaries_by_region, left_on='School Name', right_on='School Name')
joined_salaries.count()
college_region = salaries_by_college_type.merge(salaries_by_region[['School Name', 'Region']], on='School Name')
```

```python
college_region.groupby(['Region', 'School Type']).agg({"Mid-Career 25th Percentile Salary": "mean"}).round()
```
